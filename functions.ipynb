```python
import matplotlib.pyplot as plt
x = [x for x in range(1,15)]
y = [y**2 for y in x]
plt.plot(x,y)
plt.scatter(x,y,color = "red",marker="v")
plt.show()
x2 = [x for x in range(-10,10,2)]
y2 = [y2 for y2 in x2]
plt.plot(x2,y2)
plt.show()
```
