```python
X_train, X_test , y_train, y_test  = train_test_split(X,y,test_size = 0.2, random_state = 12)
```

### we need to convert the X_Train and X_test into the values between 0 and 1. This produces very best results.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
knn = KNeighborsClassifier()
```

```python
parms = {'n_neighbors' : np.arange(1,50)}
```

```python
knn_new = GridSearchCV(estimator=knn,param_grid=parms,cv = 5)
```

```python
knn_new.get_params
```

```python
knn_new.estimator
```

```python
model = knn_new.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
knn_new.best_score_
```

```python
knn_new.best_params_
```
