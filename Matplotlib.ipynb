```python
import matplotlib.pyplot as plt
x = [x**2 for x in range(90)]
plt.xlabel('x-axis')
plt.ylabel("y-axis")
plt.plot(x,'*',color = 'red')
plt.show()
x = [x**3 for x in range(-12,12,2)]
y = [y*5 for y in range(-12,12,2)]
plt.plot(x,y,'^',color = 'orange')
plt.show()
x = [x for x in range(10)]
y = [y for x in range(10)]
plt.plot(x,y)
plt.show()
plt.plot([1,3,5],[1,27,125],label= 'cubes')
plt.plot([2,4,6],[1,3,5],label= 'even-odd')
plt.legend()
plt.show()
```
