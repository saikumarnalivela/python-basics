    * I categorize this group because every college has some merit limit if that satisfy college is going to give the Admission.

```python
# Your code here
def classify(Gre):
    if Gre >= 320:
        return 1
    elif Gre > 290 and Gre < 320:
        return 2
    else:
        return 3
values = data["GRE Score"].map(classify)
data["GRE Score"] = values
print(data)
```

### I can grade the Univeristy based on the grade, 
              *If the grade is greater than or equal to 4  ---> It is high rated Univeristy"
              *If the grade is greater than 2 and less than 4 ---> It is good Univeristy
              *If the grade is less than 2 ---> it is Average Univeristy

```python
def univeristy(grade):
    if grade >= 4:
        return "High-Rated University"
    elif grade >2 and grade <4:
        return "Good University"
    else:
        return "Average Univeristy"
        
values = data["University Rating"].map(univeristy)
data["University Rating"] = values
print(data)
```

## Part 6 - Do a stratified split
Use the category group feature you created to perform a stratified split on the data.

```python
# Your code here
from sklearn.model_selection import StratifiedShuffleSplit
splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=123)
for train_indices, test_indices in splitter.split(data, data["University Rating"]):
    train_set = data.iloc[train_indices]
    test_set = data.iloc[test_indices]
print(train_set)
print(test_set)

train_set.info()
print("________________________________________________________")
test_set.info()
```
